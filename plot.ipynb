```python
import pandas as pd
df = pd.read_csv('robot_attitude.csv')

df['yaw'] = df['yaw'].str.replace(r'\D', '', regex=True)
df['pitch'] = df['pitch'].fillna(df['yaw'])
df['roll'] = df['roll'].fillna(df['yaw'])
df['yaw'] = df['yaw'].str[:3].astype(float)/100
df['pitch'] = df['pitch'].str[3:6].astype(float)/100
df['roll'] = df['roll'].str[6:9].astype(float)/100

df
df.to_csv('robot_attitudefix.csv', index=False)



```
