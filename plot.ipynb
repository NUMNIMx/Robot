```python
import pandas as pd

# ฟังก์ชันในการแยกค่าและแปลงเป็น float
def extract_values(value_str):
    values = {}
    if pd.isna(value_str):
        return values
    parts = value_str.split(', ')
    for part in parts:
        key, val = part.split(':')
        values[key] = float(val)
    return values

# ข้อมูลตัวอย่างใน DataFrame
df = df = pd.read_csv('robot_attitude.csv')

# สร้างคอลัมน์ใหม่จากข้อมูลที่แยกออกมา
extracted_values = df['data'].apply(extract_values).apply(pd.Series)

# รวมข้อมูลที่แยกออกมาเข้าไปใน DataFrame เดิม
df = pd.concat([df, extracted_values], axis=1)

# ลบคอลัมน์ที่ไม่ต้องการ
df.drop(columns=['data'], inplace=True)

print(df)
df.to_csv('atti_fix.csv', index=False)
```

```python
import pandas as pd
import re

# ฟังก์ชันในการแยกค่าและแปลงเป็น float
def extract_values(value_str):
    values = {}
    if pd.isna(value_str):
        return values
    
    # ค้นหาทุกค่าในรูปแบบ [number, number, ...]
    pattern = r'(\w+):\[(.*?)\]'
    matches = re.findall(pattern, value_str)
    
    for key, val_str in matches:
        # แปลง string เป็น list ของ float
        val_list = [float(x) for x in val_str.split(', ')]
        values[key] = val_list
    
    return values

# ข้อมูลตัวอย่างใน DataFrame
df = pd.read_csv('robot_esc.csv')

# สร้างคอลัมน์ใหม่จากข้อมูลที่แยกออกมา
extracted_values = df['data'].apply(extract_values).apply(pd.Series)

# รวมข้อมูลที่แยกออกมาเข้าไปใน DataFrame เดิม
df = pd.concat([df, extracted_values], axis=1)

# ลบคอลัมน์ที่ไม่ต้องการ
df.drop(columns=['data'], inplace=True)

print(df)
df.to_csv('esc_fix.csv', index=False)
```
