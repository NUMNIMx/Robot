```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('captured_image2.jpg')
hsl_image = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)
l_channel = hsl_image[:, :, 1]
l_channel_rescaled = (l_channel / 255.0) * 2 - 1
noise = np.random.normal(0, 1, l_channel_rescaled.shape)
combined_image = l_channel_rescaled + noise
combined_image = np.clip(combined_image, -1, 1)
combined_image = ((combined_image + 1) / 2) * 255
combined_image = combined_image.astype(np.uint8)
blurred_image = cv2.blur(combined_image, (6, 6))

plt.imshow(blurred_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('captured_image2.jpg')
hsl_image = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)
l_channel = hsl_image[:, :, 1]
l_channel_rescaled = (l_channel / 255.0) * 2 - 1
noise = np.random.normal(0, 1, l_channel_rescaled.shape)
combined_image = l_channel_rescaled + noise
combined_image = np.clip(combined_image, -1, 1)
combined_image = ((combined_image + 1) / 2) * 255
combined_image = combined_image.astype(np.uint8)
blurred_image = cv2.blur(combined_image, (6, 6))

hsl_image[:, :, 1] = blurred_image
rgb_image = cv2.cvtColor(hsl_image, cv2.COLOR_HLS2BGR)

plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
